```python
# Linear Regression Diagnostics 

import statsmodels.api as sm
import pandas as pd
```

```python
iris_dataset = pd.read_csv("iris.csv")
iris_dataset.head()
```

```python
iris_dataset

target = iris_dataset['Sepal.Length']
predictors = iris_dataset.drop(['Sepal.Length'], axis = 1)
```

```python
target
```

```python
predictors

predictors_less = iris_dataset['Petal.Width']
```

```python
predictors = pd.get_dummies(predictors)
```

```python
predictors
```

```python
model = sm.OLS(target, predictors)
results = model.fit()
results.summary()
```

```python
model_less = sm.OLS(target, predictors_less)
results_less = model_less.fit()
results_less.summary()
```
